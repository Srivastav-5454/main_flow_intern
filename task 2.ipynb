## Load the CSV file into a DataFrame:

```python
import pandas as pd
data=pd.read_csv("test.csv")
```

```python
data
```

```python
print("First few rows of the DataFrame:")
print(data.head())
```

## Handle missing values

```python
data.fillna(data.mean(), inplace=True)

# Display DataFrame info to see missing values handled
print("\nDataFrame info after handling missing values:")
print(data.info())
```

## Calculate summary statistics

```python
summary_statistics = data.describe()
print("\nSummary Statistics:")
print(summary_statistics)

# Additional summary statistics
mean_values = data.mean()
median_values = data.median()
std_dev_values = data.std()

print("\nMean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nStandard Deviation Values:")
print(std_dev_values)
```

## Filter data based on conditions

```python
filtered_data = data[data['Active'] > 100000]
filtered_data_multiple_conditions = data[(data['Active'] > 100000) & (data['Deaths'] < 1000)]

print("\nFiltered Data (Single Condition):")
print(filtered_data)

print("\nFiltered Data (Multiple Conditions):")
print(filtered_data_multiple_conditions)
```
